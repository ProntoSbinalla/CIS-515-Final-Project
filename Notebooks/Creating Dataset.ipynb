## Creating CSV File

```python
import os
import pandas as pd

rhinoplasty_dir = "../Images/Rhinoplasty"
aging_dir = "../Images/Aging Images"

data = []

# ----------------------
# RHINOPLASTY (label = 1)
# ----------------------
pairs = {}

for file in os.listdir(rhinoplasty_dir):
    if file.endswith(".jpg"):
        person = file.replace("Before.jpg", "").replace("After.jpg", "")
        
        if person not in pairs:
            pairs[person] = {}
        
        if "Before" in file:
            pairs[person]["before"] = file
        elif "After" in file:
            pairs[person]["after"] = file

for person, imgs in pairs.items():
    if "before" in imgs and "after" in imgs:
        data.append([
            person,
            imgs["before"],
            imgs["after"],
            1
        ])

# ----------------------
# AGING (label = 0)
# ----------------------
aging_pairs = {}

for file in os.listdir(aging_dir):
    if not file.endswith(".jpg"):
        continue

    # Skip files that don't follow expected pattern
    if "_" not in file:
        continue

    parts = file.split("_")

    # Need at least 3 parts: Name, Name, something (year/label)
    if len(parts) < 3:
        continue

    name = parts[0] + "_" + parts[1]  # e.g., Hanks_Tom

    if name not in aging_pairs:
        aging_pairs[name] = {}

    if "early" in file:
        aging_pairs[name]["before"] = file
    elif "older" in file:
        aging_pairs[name]["after"] = file

for person, imgs in aging_pairs.items():
    if "before" in imgs and "after" in imgs:
        data.append([
            person,
            imgs["before"],
            imgs["after"],
            0
        ])

# ----------------------
# CREATE CSV
# ----------------------
df = pd.DataFrame(data, columns=[
    "person_id", "before_image", "after_image", "label"
])

df.to_csv("../Data/dataset.csv", index=False)

print("CSV created successfully!")
```
